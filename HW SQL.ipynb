Q4

```python
%%sql
SELECT PNAME
FROM
(SELECT PNAME,count(*) as amount
FROM WORKS_ON JOIN PROJECT on WORKS_ON.PNO=PROJECT.PNUMBER
GROUP by PNAME
HAVING amount=
(SELECT max(amount)
FROM
(
SELECT PNAME,count(*) as amount
FROM WORKS_ON JOIN PROJECT on WORKS_ON.PNO=PROJECT.PNUMBER
GROUP by PNAME)))
```

Q5

```python
%%sql
SELECT FNAME
FROM
(SELECT EMPLOYEE.FNAME,count(*) as 'number of children'
FROM EMPLOYEE JOIN DEPENDENT on EMPLOYEE.SSN=DEPENDENT.ESSN
WHERE DEPENDENT.RELATIONSHIP='SON' or DEPENDENT.RELATIONSHIP='DAUGHTER'
GROUP by EMPLOYEE.FNAME
HAVING count(*)>2)
```

Q7

```python
%%sql
SELECT DNAME
FROM
(SELECT EMPLOYEE.DNUMBER,DNAME,count(*) as amount
FROM EMPLOYEE JOIN DEPARTMENT on EMPLOYEE.DNUMBER=DEPARTMENT.DNUMBER
GROUP by EMPLOYEE.DNUMBER
HAVING amount=
(SELECT max(amount)
FROM
(SELECT EMPLOYEE.DNUMBER,DNAME,count(*) as amount
FROM EMPLOYEE JOIN DEPARTMENT on EMPLOYEE.DNUMBER=DEPARTMENT.DNUMBER
GROUP by EMPLOYEE.DNUMBER)))
```

Q8

```python
%%sql
SELECT PNO as 'Project number',sum(hours) as 'total hours'
FROM WORKS_ON
GROUP by PNO
ORDER by sum(HOURS)
```

Q9

```python
%%sql
SELECT DNUMBER as 'Department number',sum(salary) as 'total salary'
FROM EMPLOYEE
GROUP by DNUMBER
ORDER by sum(SALARY) DESC
```

Q10

```python
%%sql
SELECT FNAME,LNAME
FROM EMPLOYEE
WHERE SSN in 
(SELECT SUPERSSN
FROM
((SELECT SUPERSSN,count(*) as amount
FROM EMPLOYEE
GROUP by SUPERSSN
HAVING amount = 
(SELECT max(amount)
FROM
(SELECT SUPERSSN,count(*) as amount
FROM EMPLOYEE
GROUP by SUPERSSN)))))
```

Part 2 - Titanic

```python
%%sql
select count(*)
from titanic
where fare>50 and pclass=1
```

Q2

```python
%%sql
select case
when count(*)>0 then 'there is'
else 'there is not'
end as result
from titanic
where pclass=2 and fare>(select min(fare) from titanic where pclass=1)
```

Q3

```python
%%sql
select case
when count(*)>0 then 'there is Jack on board'
else 'there is no Jack on board'
end as existence
from titanic
where name like '%Jack %'
```

Q4

```python
%%sql 
select sex,pclass,total_survived/passengers*100 as 'chances of survival'
from 
(select sex,pclass,cast(sum(survived) as float) as total_survived,cast(count(*) as float) as passengers
from titanic
group by sex,pclass
having sex<>'')
```

Part 3 - Pokemon

Q1

```python
%%sql
SELECT Pokemon.trainerID,trainername
FROM pokemon JOIN trainers on Pokemon.trainerID = Trainers.trainerID
GROUP BY Pokemon.trainerID
HAVING Pokemon.pokelevel = (SELECT MAX(pokelevel) FROM pokemon)
```

Q2

```python
%%sql
SELECT pokename
FROM Pokemon JOIN Trainers on Pokemon.trainerID=Trainers.trainerID
WHERE pokename like '%P%' AND trainername like '%A%'
```

Q3

```python
%%sql
SELECT trainername
FROM Trainers JOIN
(SELECT Pokemon.trainerID, count(*) as poke_count
FROM Pokemon JOIN Trainers on Pokemon.trainerID = Trainers.trainerID
GROUP by Trainers.trainerID
HAVING poke_count =
(SELECT max(poke_count)
FROM
(
SELECT Pokemon.trainerID, count(*) as poke_count
from Pokemon JOIN Trainers on Pokemon.trainerID = Trainers.trainerID
GROUP by Trainers.trainerID
))) as new_table on Trainers.trainerID = new_table.trainerID
```

```python
%%sql
select host_id,host_name,max(listing)
from
(select host_id,host_name,count(*) as listing
from AB_NYC_2019
where neighbourhood_group='Manhattan'
group by host_id)
```

Q2

```python
%%sql
select neighbourhood
from AB_NYC_2019
where room_type='Private room' and (price=(select min(price) from AB_NYC_2019) or price=(select max(price) from AB_NYC_2019))
group by neighbourhood
```
